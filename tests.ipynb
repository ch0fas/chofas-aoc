```python
def find_factors(n: int) -> list:
    factors = []
    for i in range(1, n+1):
        if n % i == 0:
            factors.append(i)
    
    factors.pop(0)
    #factors.pop(-1)
    
    return factors
```

```python
for k in find_factors(6):
    print(k)
```
